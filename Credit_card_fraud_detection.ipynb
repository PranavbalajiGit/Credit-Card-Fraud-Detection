Under-Sampling 

Building a sample dataset containing similar distribution of normal transactions and Fradulent Transactions

Number of Fradulent Transaction -> 492

```python
legit_sample  = legit.sample(n=492)
```

Concatenation Two Dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
new_dataset.head()
```
